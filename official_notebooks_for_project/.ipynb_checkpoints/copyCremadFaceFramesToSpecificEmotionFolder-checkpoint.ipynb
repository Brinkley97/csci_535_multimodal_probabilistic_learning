```python
def copy_face_frame_to_new_folder(path_to_faces_in_specific_folder, update_copy_location):
            
    files = os.listdir(path_to_faces_in_specific_folder)
        
    '''
    Run to move the middle frame
    '''
    files.sort()
    files_len = len(files)
    middle = int(files_len / 2)
    print("middle frame is: ", middle)
    middle_frame = files[middle]
    print(middle_frame)
        
    shutil.copy2(os.path.join(path_to_faces_in_specific_folder, middle_frame), update_copy_location)
    print("Success")
   
    '''
    Run to move all frames
    '''
    
    # for file in files:
        # shutil.copy2(os.path.join(path_to_faces_in_specific_folder, file), update_copy_location)
        # print("Success")
```

```python
def face_to_specific_emotion_folder(path_to_faces, path_to_copy_specific_face):
    """
    
    Parameters:
    path_to_faces -- str (of where to get the frames from)
    path_to_copy_specific_face -- str (of where to copy frames to)
    
    Function calls: 
    extract_features
    
    """
    
    face_folder_files = os.listdir(path_to_faces)
        
    for face_folder_file_idx in range(len(face_folder_files)):
        specific_face_folder = face_folder_files[face_folder_file_idx]
        split_specific_face_folder = specific_face_folder.split('_')
        print(face_folder_file_idx, specific_face_folder)
        
        path_to_faces_in_specific_folder = path_to_faces + specific_face_folder + "/"
        
        if 'ANG' in split_specific_face_folder:
            update_copy_location = path_to_copy_specific_face + 'ang_crema_d/'
            copy_face_frame_to_new_folder(path_to_faces_in_specific_folder, update_copy_location)
            print("Success")
        
        if 'DIS' in split_specific_face_folder:
            update_copy_location = path_to_copy_specific_face + 'dis_crema_d/'
            copy_face_frame_to_new_folder(path_to_faces_in_specific_folder, update_copy_location)
            print("Success")

        if 'FEA' in split_specific_face_folder:
            update_copy_location = path_to_copy_specific_face + 'fea_crema_d/'
            copy_face_frame_to_new_folder(path_to_faces_in_specific_folder, update_copy_location)
            print("Success")
            
        if 'HAP' in split_specific_face_folder:
            update_copy_location = path_to_copy_specific_face + 'hap_crema_d/'
            copy_face_frame_to_new_folder(path_to_faces_in_specific_folder, update_copy_location)
            print("Success")
            
        if 'NEU' in split_specific_face_folder:
            update_copy_location = path_to_copy_specific_face + 'neu_crema_d/'
            copy_face_frame_to_new_folder(path_to_faces_in_specific_folder, update_copy_location)
            print("Success")
        
        if 'SAD' in split_specific_face_folder:
            update_copy_location = path_to_copy_specific_face + 'sad_crema_d/'
            copy_face_frame_to_new_folder(path_to_faces_in_specific_folder, update_copy_location)
            print("Success")
        print()
```
