```python
# feature_extractor = ViTImageProcessor
# feature_extractor = ViTFeatureExtractor
model_name_or_path = 'google/vit-base-patch16-224-in21k'
feature_extractor = ViTFeatureExtractor.from_pretrained(model_name_or_path)
```

```python
def extract_features(path_to_video_frames, video_with_frames):
    """
    Parameters:
    path_to_video_frames -- py str (of path to each list of frames per video)
    video with frames -- py list (of frame file name per video)
    
    Return:
    extracted_features -- py
    """
    
    extracted_features_per_video = []
    
    # print(path_to_video_frames, video_with_frames)
    
    for video_with_frame in range(len(video_with_frames)):
        frame_path_with_frames = path_to_video_frames + video_with_frames[video_with_frame]
        # print(frame_path_with_frames)
        specific_frame = Image.open(frame_path_with_frames)
        image_to_extract_features_from = np.asarray(specific_frame)
        extracted_feature_per_frame = feature_extractor(image_to_extract_features_from, return_tensors='pt')
        extracted_features_per_video.append(extracted_feature_per_frame)

    return extracted_features_per_video
```

```python
def get_video_path(path_to_frames):
    """
    Parameters:
    path_to_frames -- str (of a single path to all video frames)
    feature_extractor -- transformers
    
    Function calls: 
    extract_features --
    
    Return
    """
    store_features_from_frames = []
    count_number_files_in_dir = 0
    # print(path_to_frames)
    frame_folder_files = os.listdir(path_to_frames)
    # print(frame_folder_files)
    for frame_folder_file_idx in range(len(frame_folder_files)):
        
        # print("Files in frame folder", frame_folder_file_idx, frame_folder_files[frame_folder_file_idx])
        path_to_video_frames = path_to_frames + frame_folder_files[frame_folder_file_idx] + "/"
        # print(path_to_video_frames)
        # path_to_frames = videos
        # print(path_to_frames)
        
        folder_exists = os.path.isdir(path_to_video_frames)
        # print(folder_exists)
        if folder_exists == True:
            
            list_frames_per_video = os.listdir(path_to_video_frames)
            # print(list_frames_per_video)
            # print("VIDEO FRAMES FOR: ", count_number_files_in_dir, path_to_video_frames)
            frame_features_extracted = extract_features(path_to_video_frames, list_frames_per_video)
            store_features_from_frames.append(frame_features_extracted)
            # print()
            count_number_files_in_dir += 1
        else:
            pass
    return store_features_from_frames
```
