## 1. Analysis of Variance (ANOVA)

- To compare 11 traits across 7 races
- IV: race = [black, white, east asian, south asian, hispanic, middle eastern, other]
- DV: psychological traits = [confident, egotistic, intelligent, kind, responsible, trustworthy, aggressive, caring, emotional, friendly, sociable]
- Outcome: p-value for each trait that'll represent the bias...

```python
race_series = dataset_copy.loc[0:, 'Race']
# race_series 
```

```python
race_and_traits_df = pd.concat([race_series, traits_df], axis=1)
race_and_traits_df
```

```python
# mean_of_traits_on_race = race_and_traits_df.mean()
# mean_of_traits_on_race
```

```python
def race_trait_anova(df):
    """
    df -- pd DataFrame
    """
    
    store_aov = []
    store_trait = []
    traits = traits_df.columns

    for trait in range(len(traits)):
        specific_trait = traits[trait]
#         print(specific_trait)
        store_trait.append(specific_trait)
        aov = pg.anova(dv=specific_trait, between='Race', data=race_and_traits_df, detailed=True)
#         print(aov)
        store_aov.append(aov)
#         print()
        
    return store_trait, store_aov
```

```python
# aov_p_values
aov_traits, aov_for_race_and_trait = race_trait_anova(race_and_traits_df)
```

```python
get_p_value = []
for aov_trait_idx in range(len(aov_traits)):
#     print(aov_traits[aov_trait_idx])
    get_p_value.append(aov_for_race_and_trait[aov_trait_idx]['p-unc'][0])
```

```python
# aov_traits, get_p_value
```

```python
significant_race_traits = []
significant_race_p_values = []

for p_value_idx in range(len(get_p_value)):
#     print(p_value)
    p_value = get_p_value[p_value_idx]
    
    if p_value < 0.01:
#         print(aov_traits[p_value_idx], p_value, '**')
        significant_race_traits.append(aov_traits[p_value_idx])
        significant_race_p_values.append(p_value)
    elif p_value < 0.05:
        pass
#         print(aov_traits[p_value_idx], p_value, '*')
    else:
        pass
#         print(aov_traits[p_value_idx], p_value)
```

```python
def mean_median_for_each_race(df):
    """
    """

    all_means = []
    all_medians = []
    race_trait = []
    
    race_numbers = df.loc[0:, 'Race'].unique()
    race_numbers = np.sort(race_numbers)
    
    for race_number in race_numbers:
        
        specific_race = (df['Race'] == race_number)
        specific_race_df = df.loc[specific_race]
        # print(specific_race_df)
        race_trait.append(specific_race_df)
        
        specific_race_mean = specific_race_df.mean()
        all_means.append(specific_race_mean)
        
        specific_race_median = specific_race_df.median()
        all_medians.append(specific_race_median)
        

    return all_means, all_medians, race_trait
```

- IV: gender [male, female]

```python
def female_male_ttest(f_df, m_df):
    """
    """
    
    significant_traits = []
    significant_p_values = []
    all_gender_p_values = []
    store_p_values_in_df = pd.DataFrame()
    for trait in range(len(traits)):
        specific_trait = traits[trait]
        # print(specific_trait)
        
        # print(f_df[specific_trait], m_df[specific_trait])
        stat_result, p_result = stats.ttest_ind(f_df.loc[0:, specific_trait], m_df.loc[0:, specific_trait])
        # print(stat_result)
        # print(p_result)
        
        if p_result < 0.01:
            # print(specific_trait, p_result, '**')
            significant_traits.append(specific_trait)
            significant_p_values.append(p_result)
            all_gender_p_values.append(p_result)
        elif p_result < 0.05:
            # print(specific_trait, p_result, '*')
            all_gender_p_values.append(p_result)
        else:
            # print(specific_trait, p_result)
            all_gender_p_values.append(p_result)

    return significant_traits, significant_p_values, all_gender_p_values
```

```python
significant_gender_traits, significant_gender_p_values, all_gender_trait_p_values = female_male_ttest(female_df, male_df)
# significant_gender_traits, significant_gender_p_values
```

```python
store_significant_gender_traits_with_value = {}
for significant_gender_traits_idx in range(len(significant_gender_traits)):
    trait = 'gender_' + significant_gender_traits[significant_gender_traits_idx]
    store_significant_gender_traits_with_value[trait] = significant_gender_p_values[significant_gender_traits_idx]
```
