## Preprocess Image

```python
def get_img_array(img_path, size):
    """
    
    functions
    preprocess_input -- keras vgg19
    """
    
    store_preprocess_images = []
    
    # `img` is a PIL image of size 299x299
    img = keras.preprocessing.image.load_img(img_path, target_size=size)
    
    # `array` is a float32 Numpy array of shape (299, 299, 3)
    array = keras.preprocessing.image.img_to_array(img)
    
    # We add a dimension to transform our array into a "batch"
    # of size (1, 299, 299, 3)
    array = np.expand_dims(array, axis=0)
    
    preprocess_image = preprocess_input(array)
    store_preprocess_images.append(preprocess_image)
    
    return store_preprocess_images
```

```python
store_img_arrays = []
dis_preprocessed_images_crema_d = []
# specific_dataset_frames = ang_crema_d
dis_specific_dataset_frames = dis_crema_d
for specific_dataset_frame_idx in range(len(dis_specific_dataset_frames[0:38500])):

    specific_frame_path = dis_specific_dataset_frames[specific_dataset_frame_idx]
    print(specific_dataset_frame_idx, specific_frame_path)
    
    img_array = get_img_array(specific_frame_path, size=img_size)
    dis_preprocessed_images_crema_d.append(img_array)
```

```python
def extract_features_with_vgg(store_img_arrays):
    """
    """
    
    store_extracted_features = []
    for store_img_arrays_idx in range(len(store_img_arrays)):
        store_img_array = store_img_arrays[store_img_arrays_idx]
        # print(store_img_arrays_idx, store_img_array)
        print(store_img_arrays_idx, len(store_img_array))

        features = model.predict(store_img_array)
        store_extracted_features.append(features)
    
    return store_extracted_features
```
