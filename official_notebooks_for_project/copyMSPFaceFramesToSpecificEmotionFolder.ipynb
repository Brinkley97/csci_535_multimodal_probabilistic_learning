```python
def copy_face_frame_to_new_folder(old_location_of_r_or_t, new_location_of_r_or_t):
    # print(old_location_of_r_or_t)
    files = os.listdir(old_location_of_r_or_t)
     # print(files)
    
    '''
    Run to move the middle frame
    '''
    files.sort()
    files_len = len(files)
    middle = int(files_len / 2)
    print("middle frame is: ", middle)
    middle_frame = files[middle]
    print(middle_frame)
        
    shutil.copy2(os.path.join(old_location_of_r_or_t, middle_frame), new_location_of_r_or_t)
    
   
    '''
    Run to move all frames
    '''
    # for file in files:
    #     print(file)
        
        # shutil.copy2(os.path.join(old_location_of_r_or_t, file), new_location_of_r_or_t)
```

```python
def face_to_specific_emotion_folder(path_to_faces, path_to_copy_specific_face, dataset_folder):
    """
    
    Parameters:
    path_to_faces -- str (of where to get the frames from)
    path_to_copy_specific_face -- str (of where to copy frames to)
    
    Function calls: 
    extract_features
    
    """
    
    # print(path_to_faces)
    # print(path_to_copy_specific_face)
    
    face_folder_files = os.listdir(path_to_faces)
        
    for face_folder_file_idx in range(len(face_folder_files)):
        specific_face_folder = face_folder_files[face_folder_file_idx]
        
        split_specific_face_folder = specific_face_folder[14::-16]
        print(face_folder_file_idx, specific_face_folder, '-->', split_specific_face_folder)
        
        path_to_faces_in_specific_folder = path_to_faces + specific_face_folder + "/"
        
        if 'A' in split_specific_face_folder:
            update_copy_location = path_to_copy_specific_face + 'ang' + dataset_folder
            copy_face_frame_to_new_folder(path_to_faces_in_specific_folder, update_copy_location)
            print("Success")
            
        if 'H' in split_specific_face_folder:
            update_copy_location = path_to_copy_specific_face + 'hap' + dataset_folder
            copy_face_frame_to_new_folder(path_to_faces_in_specific_folder, update_copy_location)
            print("Success")
            
        if 'N' in split_specific_face_folder:
            update_copy_location = path_to_copy_specific_face + 'neu' + dataset_folder
            copy_face_frame_to_new_folder(path_to_faces_in_specific_folder, update_copy_location)
            print("Success")
        
        if 'S' in split_specific_face_folder:
            update_copy_location = path_to_copy_specific_face + 'sad' + dataset_folder
            copy_face_frame_to_new_folder(path_to_faces_in_specific_folder, update_copy_location)
            print("Success")
        print()
```
