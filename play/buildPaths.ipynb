```python
def emotion_recording_path(df):
    """
    """
    
    emotions_paths = df.values.tolist()
    emotions_paths = sum(emotions_paths, [])
    # print(emotions_paths)
    # recording_scenarios = ["P/", "R/", "S/", "T/"]
    recording_scenarios = ["R/", "T/"]
    recording_paths = []
    r_recording_paths = []
    t_recording_paths = []

    for recording_scenario in recording_scenarios:
        for emotions_path in range(len(emotions_paths)):
            
            # print(emotions_paths[emotions_path])
            build_in_recording_path = emotions_paths[emotions_path] + recording_scenario
            recording_paths.append(build_in_recording_path)
            
            if recording_scenario == "R/": 
                r_recording_paths.append(build_in_recording_path)
                
            if recording_scenario == "T/": 
                t_recording_paths.append(build_in_recording_path)
    return recording_paths, r_recording_paths, t_recording_paths
```

```python
def emotion_recording_wav_path(recording_paths, build_base):
    """
    """
    
    # print(df)
    # recording_paths = df.values.tolist()
    # recording_paths = sum(recording_paths, [])
    # print(emotions_paths)
    
    all_rs = []
    all_ts = []
    
    
    wav_files_paths = []


    for recording_path in range(len(recording_paths)):
        directory_to_wav = build_base + recording_paths[recording_path]
        # print(directory_to_wav)
        r_or_t = directory_to_wav[-2]
        # print(r_or_t)
        path_exists = os.path.exists(directory_to_wav)
        # print(path_exists)
        
        if path_exists == True:
            # directory of wav
            wav_files = os.listdir(directory_to_wav)
            # print(wav_files)
            for wav_file in wav_files:
                build_in_wav_path = recording_paths[recording_path] + str(wav_file)
                # print(build_in_wav_path)
                wav_files_paths.append(build_in_wav_path)
                # print(wav_files_paths)
                if r_or_t == "R":
                    all_rs.append(build_in_wav_path)
                    # print(r_or_t)
                if r_or_t == "T":
                    all_ts.append(build_in_wav_path)
                    # print(r_or_t)
        else:
            pass
            # print("FILE NOT FOUND")
    
    return wav_files_paths, all_rs, all_ts
```

```python
def move_files(all_rs, all_ts):
    """
    """ 
 
    r_sources = all_rs 
    # print(r_source)
    
    r_destination = BASE + CLASS_PATH + DATASET_PATH + "r/"
    # print(r_destination)
    
    
    t_sources = all_ts 
    # print(t_source)
    t_destination = BASE + CLASS_PATH + DATASET_PATH + "t/"
    # print(t_destination)

    # gather all files
    for r_source in r_sources:
        all_r_files = os.listdir(r_source)

        # iterate on all files to move them to destination folder
        for rf in all_r_files:
            src_path = os.path.join(r_source, rf)
            dst_path = os.path.join(r_destination, rf)
            shutil.move(src_path, dst_path)
            
    # gather all files
    for t_source in t_sources:
        
        path_exists = os.path.exists(t_source)
        if path_exists == True:
            all_t_files = os.listdir(t_source)
            # iterate on all files to move them to destination folder
            for tf in all_t_files:
                src_path = os.path.join(t_source, tf)
                dst_path = os.path.join(t_destination, tf)
                shutil.move(src_path, dst_path)
        else:
            pass
```
