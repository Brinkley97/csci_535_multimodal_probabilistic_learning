# Copy MSP R and T Videos to R and T Folders Respectively

```python
import os
import os.path
import shutil

from pathlib import Path
```

```python
BASE = '/Users/brinkley97/Documents/development/'
CLASS_PATH = 'classes/csci_535_multimodal_probabilistic_learning/'
DATASET_PATH = 'datasets/project/'
```

```python
def copy_r_or_t_to_new_folder(old_location_of_r_or_t, new_location_of_r_or_t):
    # print(old_location_of_r_or_t)
    files = os.listdir(old_location_of_r_or_t)
    # print(files)
    for files_idx in range(len(files)):
        file = files[files_idx]
        print(files_idx, file)
        
        shutil.copy2(os.path.join(old_location_of_r_or_t, file), new_location_of_r_or_t)
        print("post_copy")
```

```python
def combine_rs_and_ts(path_to_msp_session, path_to_copy_specific_videos, session):
    """
    
    Parameters:
    path_to_msp_session -- str (of where to get the videos from)
    path_to_copy_specific_videos -- str (of where to copy videos to)
    
    Function calls: 
    extract_features
    
    """

    sesson_files = os.listdir(path_to_msp_session)
    # print(sesson_files)
    count_rs = []
    count_ts = [] 


    for sesson_files_idx in range(len(sesson_files)):
        specific_emotion_folders = sesson_files[sesson_files_idx]
        print(sesson_files_idx, specific_emotion_folders)

        path_to_emotion_folder = path_to_msp_session + '/' + specific_emotion_folders + '/'
        # print(path_to_emotion_folder)
        
        path_exists = os.path.exists(path_to_emotion_folder)
        # print(path_exists)
        if path_exists == True:
            
            r_t_s_p_folders = os.listdir(path_to_emotion_folder)
            print(r_t_s_p_folders, path_to_emotion_folder)        
            
            if 'R' in r_t_s_p_folders:

                old_path_to_r = path_to_emotion_folder + 'R/'
                print(r_t_s_p_folders, old_path_to_r)

                videos_in_r = os.listdir(old_path_to_r)
                
                
                count_r_videos = len(videos_in_r)
                print("Count R videos = ", count_r_videos)
                count_rs.append(count_r_videos)
                
                new_path_to_copy_r = path_to_copy_specific_videos + 'msp_r/r_videos/'
                # print(new_path_to_copy_r)
                
                copy_r_or_t_to_new_folder(old_path_to_r, new_path_to_copy_r)
            
            if 'T' in r_t_s_p_folders:

                old_path_to_t = path_to_emotion_folder + 'T/'
                print(r_t_s_p_folders, old_path_to_t)

                videos_in_t = os.listdir(old_path_to_t)
                
                
                count_t_videos = len(videos_in_t)
                print("Count T videos = ", count_t_videos)
                count_ts.append(count_t_videos)
                
                new_path_to_copy_t = path_to_copy_specific_videos + 'msp_t/t_videos/'
                # print(new_path_to_copy_t)
                
                copy_r_or_t_to_new_folder(old_path_to_t, new_path_to_copy_t)
    
            print()
    
    print()
    # print(count_r_videos)
    # print(count_t_videos)
    return count_rs, count_ts
```

```python
MSP_VIDEO_PATH = BASE + CLASS_PATH + DATASET_PATH + 'msp_videos/'
MSP_VIDEO_PATH
```

```python
COPY_MSP_PATH = BASE + CLASS_PATH + DATASET_PATH
COPY_MSP_PATH
```

```python
count_r_videos = []
count_t_videos = []

# sessions = ["session1", "session2"]
sessions = ["session1", "session2", "session3", "session4", "session5", "session6"]
for session in sessions:
 
    print(session)

    # sample_msp_with_session = SAMPLE_MSP_VIDEO_PATH + session
    # r_videos, t_videos = combine_rs_and_ts(sample_msp_with_session, SAMPLE_COPY_MSP_PATH, session)

    msp_with_session = MSP_VIDEO_PATH + session

    count_r, count_t = combine_rs_and_ts(msp_with_session, COPY_MSP_PATH, session)
    # print(count_r)
    # print(count_t)
    count_r_videos.append(count_r)
    count_t_videos.append(count_t)
    
    
```
