```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import opensmile
from HTKFeat import MFCC_HTK
from HTK import HCopy, HTKFile
```

```python
BASE = '../../csci_535_interpretable_emotion_recognition_from_audiovisual/opensmile/'
opensmile_audio_file = BASE + 'example-audio/opensmile.wav'
```

properly add to MFCC plots. Top is waveform and bottom spectrogram [4]

```python
def plot_audio_file(opensmile_audio_file):
    mfcc=MFCC_HTK()

    signal = mfcc.load_raw_signal(opensmile_audio_file)
    # print(signal)

    sig_len=signal.size/16000 #in seconds

    plt.figure(figsize=(15,4))
    t=np.linspace(0,sig_len,signal.size)
    plt.plot(t,signal)
    plt.figure(figsize=(15,4))
    s=plt.specgram(signal,Fs=16000)
    plt.xlim(0,sig_len)
```

```python
plot_audio_file(opensmile_audio_file)
```

```python
def read_mfcc12(mfcc12_file):
    """
    """
    return pd.read_csv(mfcc12_file, delimiter=";")
```
