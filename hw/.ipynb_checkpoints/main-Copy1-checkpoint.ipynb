## 1. Analysis of Variance (ANOVA)

- To compare 11 traits across 7 races
- IV: race = [black, white, east asian, south asain, hispanic, middle eastern, other]
- DV: psychological traits = [confident, egotistic, intelligent, kind, responsible, trustworthy, aggressive, caring, emotional, friendly, sociable]
- Outcome: p-value for each trait that'll represent the bias...

```python
race_numbers = dataset_copy.loc[0:, 'Race'].unique()
race_numbers = np.sort(race_numbers)
race_numbers
```

```python
race_series = dataset_copy.loc[0:, 'Race']
# race_series
```

```python
race_and_traits_df = pd.concat([race_series, traits_df], axis=1)
race_and_traits_df
```

`CORRECT or WRONG?`

```python
store_aov = []
for trait in range(1, len(traits)):
    aov = pg.anova(dv=traits[trait], between='Race', data=race_and_traits_df, detailed=True)
    store_aov.append(aov)
```

```python
gender_numbers = dataset_copy.loc[0:, 'Gender'].unique()
gender_numbers = np.sort(gender_numbers)
gender_numbers
```

```python
def filt_gen_df(df, gen_number):
    """
    Parameter:
    dataset_paths_copy -- df (entire df loaded) 
    
    Return:    
    """
    store_new_dfs = []
    filt_race = (df['Race'] == gen_number)
    race_df = df.loc[filt_race]
    # print(race_df)
    store_new_dfs.append(race_df)
    
    return store_new_dfs
```

```python

# filt_race_0 = (dataset_paths_copy['Race'] == 0)
# filt_race_0
filtered_gen_dfs = []
for gender_number in gender_numbers:
    # print(gen_number)
    filtered_gen_df = filt_gen_df(dataset_copy, gender_number)
    filtered_gen_dfs.append(filtered_gen_df)
    
```
